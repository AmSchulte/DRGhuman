```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters
import tifffile as tiff
from skimage.morphology import disk
from PIL import Image
import copy
from scipy import ndimage
from skimage import measure
from skimage.morphology import binary_dilation
import os
from tqdm.auto import tqdm
import json
from scipy import spatial
import cv2
from humanDRG import humanDRG_Iba1
```

```python
rootdir = r'G:\human DRG\tif_data\splitted'
```

```python
subdirs = []
for subdir, dirs, files in os.walk(rootdir):
    if subdir.endswith('NF_Iba1_Map2'):
        subdirs.append(subdir)
```

```python
# get values for each image for each DRG
results = []
for subdir in tqdm(subdirs):
    print(subdir)
    data = humanDRG_Iba1(subdir)
    results.append({
        "path":subdir,
        "neuronal_cell_sizes":data.neuronal_cell_sizes,
        "neurons_per_npa":data.neurons_per_npa_drg,
        "iba_per_nna":data.iba1_per_nna_drg})
```

```python
with open("Iba1_results.json", "w") as file:
    file.write(json.dumps(results, indent=4))
```
